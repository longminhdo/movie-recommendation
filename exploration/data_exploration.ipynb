```python
from ast import literal_eval
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
print(df.info())
df.head()
```

```python
genres_df = df.loc[:, ['genres']]
genres_df['genres'] = genres_df['genres'].apply(literal_eval)
genres_df = genres_df.explode('genres')

genres_list = [d.get('name') for d in genres_df.genres]
pd.Series(genres_list).value_counts(sort=False).plot(kind='bar')
```
